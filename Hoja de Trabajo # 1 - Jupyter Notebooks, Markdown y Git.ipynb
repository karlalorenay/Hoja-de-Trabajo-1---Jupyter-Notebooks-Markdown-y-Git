Resumen Version Control (GIT)
========

Idea Principal
--------

Los sistemas de control son herramientas que se utilizan para dar seguimiento a los cambios que se realizan a los codigos de programación.

Algunas de las funciones principales son:

- Mantiene el historico de cambios 
- Facilitan Colaboración 
- Mantienen la metadata como quien realizo cada cambio.

*Por que es tan util este tipo de herramienta?*

Nos permite dejar rastro de  cada uno de los cambios realizados , y tambien nos da la explicación logica del porque.

Respecto a GIT Data Model se definen los siguientes conceptos:

**SnapShops**

<i>Es el arbol de mas alto nivel para seguimiento </i>

**Seudocodigo**

Es un modelo de la historia de cambios

**Referencias**

Todas las tomas de foto pueden ser identificadas por SHA-1 hash

**Repositorio**

Esto es la data objects y las referencias

\*Conclusiones\*

* Es el sistema de control de versiones mas usado

* Hace un mapeo de todos los cambios realizados y tambien se puede reversar a la version deseada

* Hace mucho mas facil la colaboracion con otros usuarios y registra todos los cambios en un solo registro.
